```python
import pandas as pd

fund_master = pd.read_csv("../data/raw/01_fund_master.csv")
nav_history = pd.read_csv("../data/raw/02_nav_history.csv")
```

```python
fund_houses = fund_master["fund_house"].unique()

print("Number of Fund Houses:", len(fund_houses))
print("\nFund Houses:")
print(sorted(fund_houses))
```

```python
categories = fund_master["category"].unique()

print("Number of Categories:", len(categories))
print("\nCategories:")
print(sorted(categories))
```

```python
sub_categories = fund_master["sub_category"].unique()

print("Number of Sub-Categories:", len(sub_categories))
print("\nSub-Categories:")
print(sorted(sub_categories))
```

```python
risk_categories = fund_master["risk_category"].unique()

print("Number of Risk Categories:", len(risk_categories))
print("\nRisk Categories:")
print(sorted(risk_categories))
fund_master["risk_category"].value_counts()
```

```python
master_codes = fund_master["amfi_code"].nunique()
nav_codes = nav_history["amfi_code"].nunique()

print("Unique AMFI Codes in Fund Master:", master_codes)
print("Unique AMFI Codes in NAV History:", nav_codes)
```

```python
master_set = set(fund_master["amfi_code"])
nav_set = set(nav_history["amfi_code"])

missing_codes = master_set - nav_set

print("Missing Codes:", missing_codes)
print("Total Missing Codes:", len(missing_codes))
```

```python
if len(missing_codes) == 0:
    print("PASS: All AMFI codes from fund_master exist in nav_history.")
else:
    print("FAIL: Missing AMFI codes detected.")
```

```python
duplicates = fund_master["amfi_code"].duplicated().sum()

print("Duplicate AMFI Codes:", duplicates)
```
